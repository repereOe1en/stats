1. Построить график функции y1−α+β − yβ для β ∈ (0, α), где yt – квантиль распределения

(a) N(0, 1)

```r
alpha=as.double(readline())
curve(qnorm(1-alpha+x,0,1)-qnorm(x,0,1),xlim=c(0,alpha))
```

(b) Gamma(n, 1), n = 1, 2, 5, 10, 100,

```r

alpha=as.double(readline())
nmas=c(1,2,5,10,100)
for (i in (1:length(nmas))){
    n=nmas[i]
    curve(qgamma(1-alpha+x,shape=n,scale=1)-qgamma(x,shape=n,scale=1),xlim=c(0,alpha))
}
```

(c) R[0, 1],

```r
alpha=as.double(readline())
curve(qunif(1-alpha+x,0,1)-qunif(x,0,1),xlim=c(0,alpha))
```

(d) Beta(a, b), a = b = 5, a = 10, b = 2, a = 20, b = 1.

```r
alpha=as.double(readline())
amas=c(5,10,20)
bmas=c(5,2,1)
for (i in (1:3)){
    a=amas[i]
    b=bmas[i]
    curve(qbeta(1-alpha+x,a,b)-qbeta(x,a,b),xlim=c(0,alpha))
}
```
